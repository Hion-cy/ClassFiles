####**Ejercicio 01:** Errores Tipográficos y Valores Inconsistentes: maritalstatus

1.   Imprime de la columna 'maritalstatus' los valores únicos.
2. Imprime la frecuencia de cada valor único
3. Cambia el valor 'Never-Married' a 'Single'
4. Imprime la frecuencia de cada valor único.



```python
print(f"Valores unicos: {df['maritalstatus'].unique()}\n")
print(f"Frecuencia de cada valor en: {df['maritalstatus'].value_counts(dropna=False)}\n") #Agregue el dropna=false por que hay faltan valores
df['maritalstatus'] = df['maritalstatus'].replace('Never-married', 'Single')
print(f"Frecuencia de cada valor en: {df['maritalstatus'].value_counts(dropna=False)}\n")
```

####**Ejercicio 02:** Errores Tipográficos y Valores Inconsistentes: highestdegree.
1. Imprime de la columna 'highestdegrees' los valores únicos.
2. Imprime la frecuencia de cada valor único
3. Observa que cada valor categórico empieza con 0. Modifica las categorías para que todas estén en minúsculas y ninguna empiece con un número. Por ejemplo, si dice "2. High School", debería ser 'high school.
4. Los cambios se realizarán en la misma columna 'highestdegrees'.
5. Imprime los nuevos valores únicos.

```python
print(f"Valores unicos: {df['highestdegree'].unique()}\n")
print(f"Frecuencia de cada valor en: {df['highestdegree'].value_counts(dropna=False)}\n") #Agregue el dropna=false por que hay faltan valores
df['highestdegree'] = df['highestdegree'].str.replace(r"[0-9]+\.\s", "", regex=True).str.lower()
print(f"Nuevos valores unicos: {df['highestdegree'].unique()}\n")
```

####**Ejercicio 03:** Alta Cardinalidad - Frecuencias Bajas

1. De la columna 'maritalstatus' une las categorías Widowed y 'Single' en una categoría más general llamada 'Single/No Partner'. Crea un|a nueva columna, 'marital_grouped' donde aparezcan las nuevas categorías de maritalstatus.[texto del enlace](https://)
2. Imprime los valores únicos de la nueva columna 'marital_grouped'.
3. Imprime la frecuencia de los valores únicos de la nueva columna 'marital_grouped'

```python
df['marital_grouped']=df['maritalstatus'].replace(['Widowed','Single'],'Single/No Partner')
print(f"Valores unicos: {df['marital_grouped'].unique()}\n")
print(f"Frecuencia de cada valor en: {df['marital_grouped'].value_counts(dropna=False)}\n") #Agregue el dropna=false por que hay faltan valores
```

####**Ejercicio 04:** Tipos de Datos

1. Imprime el tipo de datos de la columna gender
2. Imprime la cantidad de bytes de dicha columna
3. Si la columna es de tipo object, modifica a category.
4. En caso de haber realizado la modificación del tipo de datos, imprime la cantidad de bytes que ocupa dicha columna con el nuevo tipo de datos

```python
print(f"Tipo de dato de 'Gender': {df['gender'].dtype}\n")
print(f"Cantidad de bytes: {df['gender'].nbytes}")
if df['gender'].dtype == 'object':
  print("El tipo de datos es object y sera actualizado a category")
  df['gender'] = df['gender'].astype('category')
  print(f"La nueva cantidad de bytes es de: {df['gender'].nbytes}\n")
else:
  print("El tipo de datos ya es category")
```
