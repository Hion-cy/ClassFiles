###Ejercicio 1

#### Transformación
* Crear un CountVectorizer y un TfidfVectorizer.
* Aplícalos al corpus y transformar los textos.

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
```

```python
vectorizer= CountVectorizer()
X=vectorizer.fit_transform(corpus_es)


bow_df= pd.DataFrame(
    X.toarray(),
    columns=vectorizer.get_feature_names_out(),
    index=[f"d{i+1}" for i in range(len(corpus_es))]
)

#print(X)
#print(X.toarray(),"\n")
#print(list(bow_df.columns))
#print(vectorizer.get_feature_names_out())
print("---- CountVectorizer ----")
print(bow_df,"\n")
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus_es) # Convertir a DataFrame
Tfid_df = pd.DataFrame(
     X.toarray(),
     columns=vectorizer.get_feature_names_out() )
print("---- TfidfVectorizer ----")
print(Tfid_df, "\n")
```

####Mostrar

1.	Dimensiones de la matriz resultante.
2.	Primeros 10 tokens y sus índices.


```python
print(f"Dimensiones BoW (Filas, Columnas): {bow_df.shape}")
print(f"Dimensiones TF-IDF (Filas, Columnas): {Tfid_df.shape}")
```

```python
tokens = vectorizer.get_feature_names_out()
print("---- Primeros 10 tokens y sus índices ----")
for i in range(10):
    print(f"Índice {i}: {tokens[i]}")
```

####Comparar las diferencias entre ambas matrices (CountVectorizer y TfidfVectorizer), mencionando de manera general las diferencias observadas.

**Conclusión:** Al no integrar ningún parámetro en la realizacion de CountVectorizer y TfidfVectorizer el tamaño de ambas matrices se mantiene igual para ambas transformaciones (30x127).

Sin embargo existen palabras como "aburrida" donde, si solo tomáramos en cuenta el primero (Count) no sería muy útil ya que en casos como d4 y d21 ambos poseen el valor de 1, mientras que en el segundo (Tfid) el peso varía entre 0.460672 y 0.470578 donde se puede afirmar que la palabra tiene más peso en la segunda reseña (d21 o 20).

Otros ejemplos como actuación (d6 (5) y d22 (21)) refuerzan la diferencia entre emplear un método u otro para valorar la importancia de cada palabra.


### Ejercicio 2

Preprocesa el texto eliminando los acentos y stopwords, así como convirtiéndolo a minúsculas.

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words_es = stopwords.words('spanish')
```

#### Realiza las mismas actividades que en el ejercicio anterior.

```python

vectorizer_parametros= CountVectorizer(lowercase=True, stop_words=stop_words_es, strip_accents='unicode' )
X_parametros=vectorizer_parametros.fit_transform(corpus_es)

print("---- CountVectorizer ----")
bow_param= pd.DataFrame(
    X_parametros.toarray(),
    columns=vectorizer_parametros.get_feature_names_out(),
    index=[f"d{i+1}" for i in range(len(corpus_es))]
)

print(bow_param,"\n")
```

```python

vectorizer_parametros= TfidfVectorizer(lowercase=True,  stop_words=stop_words_es, strip_accents='unicode' )
X_parametros=vectorizer_parametros.fit_transform(corpus_es)

Tfid_param = pd.DataFrame(
     X_parametros.toarray(),
     columns=vectorizer_parametros.get_feature_names_out() )
print("---- TfidfVectorizer ----")
print(Tfid_param, "\n")
```

```python
print(f"Dimensiones BoW (Filas, Columnas): {bow_param.shape}")
print(f"Dimensiones TF-IDF (Filas, Columnas): {Tfid_param.shape}")
```

```python
tokens_param = vectorizer_parametros.get_feature_names_out()
print("---- Primeros 10 tokens y sus índices ----")
for i in range(10):
    print(f"Índice original {i+1}: {tokens[i]}")
    print(f"Índice {i+1}: {tokens_param[i]}\n")
```

####	Compara las matrices del ejercicio anterior con las matrices que obtuviste en este ejercicio, mencionando de manera general las diferencias observadas.

**Conclusión:** La diferencia principal entre las matrices originales y las matrices "limpias" es que al eliminar las stop-words el tamaño se reduce de 30x127 a 30x97.

Además al eliminar los acentos se unifican palabras como "acción" y "acción" para interpretarse como la misma palabra permitiendo que el conteo y peso se repartan de manera más uniforme.

Los índices también cambian, en el caso del índice 6 que originalmente era "al" paso a ser "ambientación" (en el original ocupa el lugar 8) que da un contexto más acercado al tema (películas). Lo mismo ocurre con "amigos" que paso del lugar 9 al 7.
