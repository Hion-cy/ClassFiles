## Instrucciones


###1. Carga de datos
Carga el dataset Titanic usando spark.read.csv

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("TitanicData").getOrCreate()
df = spark.read.csv('/content/drive/MyDrive/ClassFiles/titanic.csv', header=True, inferSchema=True)
```

####Comentarios adicionales:

Se inicializa el dataset después de haber sido cargado en el folder de Google drive. Adicional a solo leer el csv, se utilizo la opción header=True para indicar que el origen de los datos incluye encabezados asi como inferSchema=True para detectar los tipos de datos.

###2. Exploración inicial
* Muestra las primeras filas con show()
* Visualiza el esquema con printSchema()


```python

print("\n Visualiza el contenido usando show ")
df.show()

print("\n Visualiza del schema ")
df.printSchema()
```

####Comentarios adicionales:


La función show() nos permite observar una vista previa del contenido (por defecto los primeros 20), mientras que printSchema() muestra la configuracion de cada columna.

###3. Selección de columnas
* Selecciona las columnas: Name, Age, Sex
* Muestra el resultado


```python
cols_selected =['Name','Age','Sex']
print("Columnas seleccionadas")
#print(df.select(cols_selected).show())
df.select(cols_selected).show()
```

####Comentarios adicionales:

Para seleccionar varias columnas se puede crear primero una "lista" con las que se busca utilizar, para posteriormente usarlas como argumento en la función select() para poder ejecutarla sobre el DataFrame original.

###4. Filtrado de datos
1. Filtra los pasajeros:
* Mayores de 18 años
* De sexo femenino
2. Muestra los resultados

```python
print("Mayores de 18 años")
df.filter((df.Age > 18)).show()
#df.filter((df.Age > 18)).select(cols_selected).show(
```

```python
print("Sexo femenino")
df.filter((df.Sex == 'female')).show()
#df.filter((df.Sex == 'female')).select(cols_selected).show()
```

```python
print("Mayores de 18 años y sexo femenino")
df.filter((df.Age > 18) & (df.Sex == 'female')).show()
```

####Comentarios adicionales:

Se aplicaron filtros para valores  numericos (>) y cadenas de texto ("female") para obtener los resultados solicitados. Asi como hacer una ultima consulta utilizando & para unir ambos filtros en una sola instrucción.

###5. Valores únicos
* Obtén los valores únicos de la columna Pclass

```python
print("Valores únicos de la columna Pclass")
df.select('Pclass').distinct().show()
```

####Comentarios adicionales:

La función distinct() permite identificar los valores unicos de un campo dado sin necesidad de filtros adicionales. En la actividad actual es posible observar como los pasajeros del Titanic estaban divididos en 3 clases.

###6. Agrupación
Cuenta el número de pasajeros por:
* Pclass
* Sex


```python
print("Cuenta el número de pasajeros por clase")
df.groupBy('Pclass').count().show()
```

```python
print("Cuenta el número de pasajeros por Sexo")
df.groupBy('Sex').count().show()
```

####Comentarios adicionales:

Para obtener el numero de pasajeros por sexo y clase, se realiza primero la agrupacion para después utilizar la función count() para obtener el resultado deseado.

###7. Estadística básica
Aplica describe() sobre las variables numéricas


```python
numeric_var=['Age','Fare','Siblings/Spouses Aboard','Parents/Children Aboard']
#numeric_var=['Age','Fare']
df.select(numeric_var).describe().show()
```

####Comentarios adicionales:
El método describe() proporciona una vista previa de los valores estadisticos de las variables numericas del DataFrame, esto permite detectar si existen rangos inesperados en los datos.

###8. Valores nulos
Identifica en qué columnas hay valores nulos (usa isNull() o funciones vistas)

```python
print("Valores nulos")
for col in df.columns:
  print(col, df.filter(df[col].isNull()).count())
#
```

####Comentarios adicionales:

Se utilizo un ciclo de iteración para recorrer todas las columnas del DataFrame para validar la integridad de los datos.
Los resultados demuestran que ninguna columna cuenta con datos nulos que deban ser contabilizados.
