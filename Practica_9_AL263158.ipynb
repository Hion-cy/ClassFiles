```python
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

df = px.data.gapminder()
```

```python
df2007 = df[df['year'] == 2007]
#Grafico de dispersion con plotly.express
colors={"Asia":"red","Europe":"yellow","Africa":"brown","Americas":"blue","Oceania":"purple"}
fig=px.scatter(df2007, x='gdpPercap', y='lifeExp', size='pop', color='continent', hover_name='country', log_x=True, size_max=60, labels={
        'gdpPercap': 'GDP per capita',
        'lifeExp': 'Life Expectancy',
        'continent': 'Continent',
        'pop': 'Population'
    },
    title='GDP per capita vs Life Expectancy (2007)',
    color_discrete_map=colors
)

fig.show()


```

```python

#Grafico de dispersion seaborn
plt.figure(figsize=(10,6))

sns.scatterplot(data=df2007, x='gdpPercap', y='lifeExp', hue='continent',palette=colors,size="pop", sizes=(50, 500), alpha=0.5)

plt.xlabel('GDP per capita')
plt.ylabel('Life Expectancy')
plt.tight_layout()
plt.show()
```
