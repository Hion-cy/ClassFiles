```python
#Matricula AL263158
#Programación para analítica descriptiva y predictiva

from google.colab import files
from google.colab import drive
import shutil
import os

drive.mount('/content/drive')

print(f"Bienvenido al analisis de archivos de acciones\n",
      "Por favor siga las siguientes instrucciones para subir el archivo a la plataforma\n",
      "1. Utilize el comando \"drive.mount('/content/drive')\" para montar Google Drive, si ya lo hizo previamente omita este paso\n ",
      "2. Seleccione \"Elegir archivos\" para elegir el archivo deseado",
      "3. Escriba la ruta deseada en Drive\n4. Presione \"Enter\"")

print("\nEjemplo\n1. \"drive.mount('/content/drive')\"\n2. Seleccionar archivo \"informe_acciones.txt\"\n3. Ruta = \"/content/drive/MyDrive/ClassFiles/\"\n\n")



uploaded = files.upload()
carpeta_destino = input("Ingrese la ruta deseada: ")

try:
  for fn in uploaded.keys():
    ruta_final = os.path.join(carpeta_destino, fn)
    shutil.move(fn, carpeta_destino)
except FileNotFoundError:
  print("Ruta no encontrada")


```

```python
import re
print(ruta_final)
archivo = open(ruta_final, "r", encoding="latin-1")
tabla=[]
cont=0
for linea in archivo:
    linea = linea.rstrip()
    #print(linea)
    nombre=re.findall('^Te escribo para informarte sobre el valor actual de la acción de (.+?)\s\(',linea)
    #^Te escribo para informarte sobre el valor actual de la acción de... El mensaje se repite en cada linea de correo
    # Quiero obtener unicamente lo que va despues de eso (Uno o mas caracteres de manera no ambisiosa)
    # Todas los mensajes tienen la abreviatura con un parentesis al final, vamos a extraer el nombre hasta antes del espacio en blando y el parentesis
    fecha=re.findall(r'Fecha: ([0-9]{4}-[0-9]{2}-[0-9]{2})',linea)
    #Busca la fecha junto con su formamto 4 numeros para el año, - , 2 para el mes, - , 2 para el dia
    valor=re.findall(r'Valor actual: \$([0-9.]+)',linea)
    #El valor va acompañado de un espacio y un signo de $ y puede ser mas de un digito por eso +

    if len(nombre) != 0:
      #print(nombre[0])
      nombre_app=nombre[0]
      cont=cont+1
    elif len(fecha) != 0:
      #print(fecha[0])
      fecha_app=fecha[0]
      cont=cont+1
    elif len(valor)!=0:
      #print(valor[0])
      valor_app=valor[0]
      cont=cont+1
      #print(nombre_app,fecha_app,valor_app)
    else:
      cont= 0
    if cont==3: # Siempre estan los 3 renglones juntos por ello se agregan de 3 en 3 los valores en una matriz
      tabla.append([nombre_app,fecha_app,valor_app])
      cont=0

resumen=[]
fechareciente=""
v_max=0
v_min=0
prom=0
empresas = set(columnas[0] for columnas in tabla) #Separarlos valores unicos de los nombres
for empresa in empresas:
  #valor= [columnas[2] for columnas in tabla if columnas[0] == empresa] #De la columna 2 (Valor), se agrega al arreglo valor para cada valor de la tabla si la comlumna 0 = empresa del for
  valor= [float(columnas[2]) for columnas in tabla if columnas[0] == empresa]
  fecha= [columnas[1] for columnas in tabla if columnas[0] == empresa]
  fechareciente=max(fecha)
  v_max=max(valor)
  v_min=min(valor)
  prom=round(sum(float(i) for i in valor)/len(valor),2)

  resumen.append([empresa,fechareciente,v_max,v_min,prom])

print(f"{'Empresa':40} {'Fecha más reciente':15} {'Valor Máx ($)':15} {'Valor Mín ($)':15} {'Promedio ($)':15}")
for empresa in resumen:
  print(f"{empresa[0]:40} {empresa[1]:15} {empresa[2]:15} {empresa[3]:15} {empresa[4]:15}")


```
